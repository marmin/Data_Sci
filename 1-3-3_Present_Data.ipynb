Choose one variable and plot that variable four different ways

```python
df_IRN = dfn[dfn.Entity == 'Iran']
sns.set_style('white')
ax = sns.factorplot(x='Year', y='Total population (Gapminder)', data=df_IRN, size=10)
plt.xticks(rotation=45)
plt.show()
```

```python
ax = sns.factorplot(x='Year', y='Total population (Gapminder)', data=df_IRN, size=10, kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python
ax = sns.boxplot(y='Total population (Gapminder)', data=df_IRN)
sns.despine(offset=10, trim=True)
```

```python
#point plot
g = sns.factorplot( y='Total population (Gapminder)', 
                   data=df_IRN, kind='point', size=10, dodge=True)
g.despine(left=True)
plt.xticks(rotation=45)
plt.show()
```

- Probability distributions, 
- Bivariate relationships, 
- Whether the distributions or the relationships vary across groups

```python
df_IRN_FR = df[(df.Entity =='Iran') | (df.Entity =='France')]
df_IRN_FR = df_IRN_FR.dropna()
df_IRN_FR = df_IRN_FR.rename(columns={'GDP per capita (2011 int-$) (US$ (PPP-adjusted; real 2011))':'GDP per capita', 
                       'Daily caloric supply (kcal/person/day)':'Daily caloric supply',
                      'Total population (Gapminder)':'Total population'})
df_IRN_FR.head(2)
```

```python
x = df_IRN_FR[df_IRN_FR['Entity']=='Iran']['Daily caloric supply']
y = df_IRN_FR[df_IRN_FR['Entity']=='France']['Daily caloric supply']
g = sns.distplot(x, kde=False)
g1 = sns.distplot(y, kde=False)
plt.show()
x = df_IRN_FR[df_IRN_FR['Entity']=='Iran']['GDP per capita']
y = df_IRN_FR[df_IRN_FR['Entity']=='France']['GDP per capita']
g = sns.distplot(x, kde=True)
g1 = sns.distplot(y, kde=True)
plt.show()
```
